## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

Flying to China and to Italy are mutually exclusive, so the probability of getting Covid is given by:

$$ P(\text{Covid}) = P(\text{Covid} | \text{China}) \cdot P(\text{China}) + P(\text{Covid} | \text{Italy}) \cdot P(\text{Italy}) $$

Substituting the values, we get:

$$ P(\text{Covid}) = \frac{1}{24} $$

And we know by Bayes' theorem:

$$ P(\text{China} | \text{Covid}) = \frac{P(\text{Covid} | \text{China}) \cdot P(\text{China})}{P(\text{Covid})} $$

So,

$$ P(\text{China} | \text{Covid}) = 0.20 $$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

n_simulations = 100000

# Simulate rolling two dice
dice_rolls = np.random.randint(1, 7, size=(n_simulations, 2))
sums = np.sum(dice_rolls, axis=1)

# Count the occurrences of sums 7, 8, and 9
sum_counts = np.bincount(sums, minlength=13)

# Extract counts for sums 7, 8, and 9
counts = sum_counts[7:10]
sum_labels = [7, 8, 9]
probs = counts/n_simulations

# Plotting the results
plt.figure(figsize=(10, 6))
sns.barplot(x=sum_labels, y=counts)
plt.title('Count of Sums 7, 8, and 9')
plt.xlabel('Sum')
plt.ylabel('Count')

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=sum_labels, y=probs)
plt.title('Count probability of Sums 7, 8, and 9')
plt.xlabel('Sum')
plt.ylabel('Prob')

# Annotating the bars with the counts
for i in range(len(probs)):
    ax.text(i, probs[i], probs[i], ha='center', va='bottom')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the seed for reproducibility
np.random.seed(42)

# Number of simulations
n_simulations = 10000

# Simulate rolling two dice
dice_rolls = np.random.randint(1, 7, size=(n_simulations, 2))
sums = np.sum(dice_rolls, axis=1)

# Count the occurrences of sums 7, 8, and 9
sum_counts = np.bincount(sums, minlength=13)

# Extract counts for sums 7, 8, and 9
counts = sum_counts[7:10]
sum_labels = [7, 8, 9]

# Plotting the results
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=sum_labels, y=counts)
plt.title('Count of Sums 7, 8, and 9 in 10,000 Dice Rolls')
plt.xlabel('Sum')
plt.ylabel('Count')

# Annotating the bars with the counts
for i in range(len(counts)):
    ax.text(i, counts[i] + 50, counts[i], ha='center', va='bottom')

plt.show()
```
